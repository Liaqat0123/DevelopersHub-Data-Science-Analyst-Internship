Importing Libraries pandas Matplotib seaborn 

```python
import pandas as pd
import matplotlib.pyplot as Plt
import seaborn as sns  
```

Load iris Dataset using pandas

```python
path="https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df=pd.read_csv(path)

df.head()
```

Display Dataset Structure


```python
#Show the Shape of Dataset
print("Shape of Dataset:")
print(df.shape)

#Show the Columns of Dataset
print("\nColumns of Dataset:")
print(df.columns)

#Show the First five rows of dataset 
print("\n First Five rows of Dataset:")
print(df.head())
```

Dataset Information

```python
#Here we describe the information
print("Dataset info:")
print(df.info())

#Dataset summary
print("\nDataset Summary:")
print(df.describe())
```

Visualization Of Dataset using Scatter, Histogram and Box plot

Scatter Plot

```python
#In the Scatter plot we were analyzing the relationships between variables. 


Plt.figure(figsize=(8,5))

sns.scatterplot(
    data=df,
    x='sepal_length',
    y='petal_length',
    hue='species'
)

Plt.title("Scatter Plot: Sepal Length vs Petal Length")

Plt.show()
```

Histogram plot

```python
#Histogram to examine the data distribution

Plt.Figure(figsize=(8,5))
sns.histplot(df['sepal_length'])
Plt.title("Histogram of sepal length")
Plt.show()
```

Box Plot

```python
# Box plot to detect outliers and spread of values. 
Plt.Figure(figsize=(8,5))
sns.boxplot(
    data=df,
    x='species',
    y='sepal_length'
)

Plt.title("Box Plot: Sepal Length by Species")
Plt.show()
```
